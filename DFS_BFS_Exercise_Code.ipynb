Here is a sample graph. We can add our customize graph with branching factor and depth. Try exercise example here.


```
graphs = {'A': set(['B', 'C']),
         'B': set(['A', 'D', 'E']),
         'C': set(['A', 'F', 'G']),
         'D': set(['B']),
         'E': set(['B']),
         'F': set(['C']),
         'G': set(['C'])}

```

```
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    for next in graph[start] - visited:
        dfs(graph, next, visited)
    return visited

dfs(graphs, 'A') # {'E', 'D', 'F', 'A', 'C', 'B'}
```

```
def dfs_paths(graph, start, goal):
    stack = [(start, [start])]
    while stack:
        (vertex, path) = stack.pop()
        for next in graph[vertex] - set(path):
            if next == goal:
                yield path + [next]
            else:
                stack.append((next, path + [next]))

list(dfs_paths(graphs, 'A', 'G')) # [['A', 'C', 'F'], ['A', 'B', 'E', 'F']]
```

```
def bfs(graph, start):
    visited, queue = set(), [start]
    while queue:
        vertex = queue.pop(0)
        if vertex not in visited:
            visited.add(vertex)
            queue.extend(graph[vertex] - visited)
    return visited

bfs(graphs, 'A') # {'B', 'C', 'A', 'F', 'D', 'E'}
```

```
def bfs_paths(graph, start, goal):
    queue = [(start, [start])]
    while queue:
        (vertex, path) = queue.pop(0)
        for next in graph[vertex] - set(path):
            if next == goal:
                yield path + [next]
            else:
                queue.append((next, path + [next]))

list(bfs_paths(graphs, 'A', 'F')) # [['A', 'C', 'F'], ['A', 'B', 'E', 'F']]
```
